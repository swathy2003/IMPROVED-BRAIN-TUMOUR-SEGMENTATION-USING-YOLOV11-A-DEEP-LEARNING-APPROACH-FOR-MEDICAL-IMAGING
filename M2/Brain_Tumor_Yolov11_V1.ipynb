```python
import pandas as pd

# Path to the results.csv file
csv_file_path = "/content/runs/segment/train/results.csv"  # Replace with your actual file path

# Load the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

# List of metrics you want to find the highest values for
metrics = [
    'metrics/precision(B)',
    'metrics/recall(B)',
    'metrics/mAP50(B)',
    'metrics/mAP50-95(B)',
    'metrics/precision(M)',
    'metrics/recall(M)',
    'metrics/mAP50(M)',
    'metrics/mAP50-95(M)'
]

# Check if the required columns exist in the file
missing_columns = [col for col in metrics if col not in df.columns]
if missing_columns:
    print(f"The following columns are missing in the CSV file: {missing_columns}")
else:
    # Get the highest value for each metric
    print("Highest values for each metric:")
    for metric in metrics:
        highest_value = df[metric].max()
        print(f"{metric}: {highest_value:.4f}")
```
